**Showing idea and concept about numpy ndarray**

```python
# import packages and using modules
import numpy as np
import matplotlib.pyplot as plt

# Generate data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

plt.plot(t, s)
plt.grid()
plt.show()
```

```python
print("type(t):", type(t))
print(t)
```

```python
print("type(s):", type(s))
print(s)
```

```python
# put t and s values side-by-side in two columns
np.hstack((t.reshape(-1,1), s.reshape(-1,1)))
```
